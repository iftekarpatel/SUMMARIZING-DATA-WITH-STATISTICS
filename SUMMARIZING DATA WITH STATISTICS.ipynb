```python
newdf = df[['LotArea','SalePrice']].copy()

# Calulating Pearson correlation coefficient
covariance = newdf.cov().iloc[0,1]
std_LotArea = newdf.LotArea.std()
std_SalePrice = newdf.SalePrice.std()
pearson = covariance/(std_LotArea*std_SalePrice)
print("Pearson's Correlation Coefficient : ", pearson)

# Calculating Spearman rank correlation coefficient
ranks = newdf.rank(axis=0)
ranks['d^2'] = (ranks['LotArea'] - ranks['SalePrice'])**2
d_square = ranks['d^2'].sum()
n = ranks.shape[0]
spearman = 1 - ((6*d_square)/(n**3 - n))
print("Spearman's Rank Correlation Coefficient : ", spearman)
```
